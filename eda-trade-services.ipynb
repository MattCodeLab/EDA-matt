```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the' dataset
trade_services = pd.read_csv('trade_services.csv')
```

```python
# convert the columns into billions from millions
trade_services['total_imports'] = trade_services['total_imports'] / 1000
trade_services['total_exports'] = trade_services['total_exports'] / 1000
trade_services['net'] = trade_services['net'] / 1000
```

```python
# plot all the data on the same axis

fig, ax1 = plt.subplots(figsize=(6, 6))

ax1.bar(trade_services['year'], trade_services['net'],
         color = '#cecece', label='Net')

ax1.set_title('Trade Services: FY2019-2024', fontsize=14, color='black')

ax1.set_ylabel('Net Position (RM bil)', color='black')

ax2 = ax1.twinx()

ax2.plot(trade_services['year'], trade_services['total_imports'], 
         color='red', marker='o', linewidth=2, label='Total Imports')
ax2.plot(trade_services['year'], trade_services['total_exports'], 
         color='green', marker='o', linewidth=2, label='Total Exports')

ax2.set_ylabel('Imports / Exports Services (RM bil)', color='black')

# add legends
bars1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(bars1 + lines2, labels1 + labels2, loc='lower right')

# Set chart box color for both axes
for spine in ax1.spines.values():
    spine.set_edgecolor('#cecece')

for spine in ax2.spines.values():
    spine.set_edgecolor('#cecece')


# Add grid
# ax2.grid(True, alpha=1, linestyle='--')
```
